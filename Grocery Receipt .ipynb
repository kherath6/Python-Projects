```python
from datetime import datetime


# [ ] The `items` list contains information about a purchase transaction at a grocery store.
# Each item in the list is another list containing:
#    1) Item description
#    2) Price per item
#    3) Quantity purchased

# Write a program to generate a receipt similar to the one below:
# 1) The first line should contain the current date and time (HINT: use `datetime` and `strftime`)
# 2) Second line should be a dashed line
# 3) Each of the items will be displayed on 2 lines:
#     I) First line contains: item number, description, total item price (based on the quantity)
#    II) Second line contains: quantity purchased, followed by price per item
# 4) Line before last should be a dashed line
# 5) Last line should contain the word TOTAL followed by the receipt total, that you should calculate


#Sun December 12, 2017 @ 04:56 P 
#--------------------------------
#
#1 - APPLES 1LB          $ 3.98
#    2.0 @ $ 1.99
#2 - OLIVE OIL           $10.99
#    1.0 @ $10.99
#3 - TOMATOS 1LB         $ 3.35
#    2.6 @ $ 1.29
#4 - MILK 1/2G           $ 3.45
#    1.0 @ $ 3.45
#5 - FLOUR 5LB           $ 2.99
#    1.0 @ $ 2.99
#6 - BELL PEPPERS 1LB    $ 3.78
#    2.8 @ $ 1.35
#7 - WHITE TUNA          $ 1.69
#    1.0 @ $ 1.69
#8 - CHEESE 1/2LB        $ 9.98
#    2.0 @ $ 4.99
#--------------------------------
#                 TOTAL $ 40.21
```

```python


items = [["APPLES 1LB", 1.99, 2], ["OLIVE OIL", 10.99, 1], ["TOMATOS 1LB", 1.29, 2.6], 
         ["MILK 1/2G", 3.45, 1], ["FLOUR 5LB", 2.99, 1], ["BELL PEPPERS 1LB", 1.35, 2.8], 
         ["WHITE TUNA", 1.69, 1], ["CHEESE 1/2LB", 4.99, 2]]

#TODO
total = 3.99 + 10.99 + 3.35 + 1.73 + 14.95 + 3.78 + 1.69 + 2.50

dt = datetime(day = 12 ,month =12, year =2017,hour = 16, minute = 56)
dt1 = dt.strftime('%a %B %d, %Y @ %I,%M,%p')


print(dt1)
print('--------------------------------------------')

print('1 - ',items[0][0],'{:>25s}{:>3.2f}'.format('$',(1.99 * 2)))
print('{:9.1f} @ {:2.2f}'.format(items[0][2],items[0][1]))

print('2 - ',items[1][0],'{:>25s}{:>4.2f}'.format('$',items[1][1]))
print('{:9.1f} @ {:2.2f}'.format(items[1][2],items[1][1]))

print('3 - ',items[2][0],'{:>24s}{:>4.2f}'.format('$',(1.29 * 2.6)))
print('{:9.1f} @ {:2.2f}'.format(items[2][2],items[2][1]))

print('4 - ',items[3][0],'{:>26s}{:>4.2f}'.format('$',(3.45/2)))
print('{:9.1f} @ {:2.2f}'.format(items[3][2],items[3][1]))

print('5 - ',items[4][0],'{:>25s}{:>4.2f}'.format('$',(2.99 * 5)))
print('{:9.1f} @ {:2.2f}'.format(items[4][2],items[4][1]))

print('6 - ',items[5][0],'{:>19s}{:>4.2f}'.format('$',(2.8 * 1.35)))
print('{:9.1f} @ {:2.2f}'.format(items[5][2],items[5][1]))

print('7 - ',items[6][0],'{:>25s}{:>2.2f}'.format('$',items[6][1]))
print('{:9.1f} @ {:2.2f}'.format(items[6][2],items[6][1]))

print('8 - ',items[7][0],'{:>23s}{:>2.2f}'.format('$',(4.99 / 2)))
print('{:9.1f} @ {:2.2f}'.format(items[7][2],items[7][1]))

print('----------------------------------------------')
print('{:>39s} {:2.2f}'.format('Total $',total))
```
